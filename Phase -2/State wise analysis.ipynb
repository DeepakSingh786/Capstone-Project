```python
import pandas as pd # used to access the values present in our data
import numpy as np # Used to access the numerical values
import matplotlib.pyplot as plt #This library used for the visualisation data in graphical form
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
state_area = pd.read_csv('Area.csv')
state_area.head()
```

```python
population = pd.read_csv('Population.csv')
population.head()
```

```python
literacy_rate = pd.read_csv('Literacy_rate.csv')
literacy_rate.head()
```

```python
literacy_rate.sort_values(by="literacy_rate", ascending=False)
```

```python
crimes = pd.read_csv('total_crimes.csv')
crimes.head()
```

# Literacy Rate vs Total Crimes

```python
req_df = pd.merge(crimes ,literacy_rate, on=['state'])
req_df["%age cases"] = (req_df["total_crimes"] / 12)*100
req_df
```

```python
req_df.sort_values(by="literacy_rate", ascending=False)
```

Observations --

1. Most literate state is Kerala
2. Unlikely Kerala is considering higher number of cases as compared to the other literate states followed by Maharashtra and Delhi
3.As we can see, there is an upward trend to be observed between literacy rates and the number of cases
4. Surprisingly, the least literate states perform much better, with Arunachal Pradesh and Bihar quoting only 1 to 5% of cases respectively. 
5. On the other hand, these low percentages in the lesser literate states of India could also point to the fact that a lack of formal education may deter the common man from approaching the judicial system

```python
fig, ax = plt.subplots(figsize=(30,10))

req_df.plot(x="state", y="literacy_rate", ylabel = "cases as % of total", ax=ax)
req_df.plot(x="state",y="%age cases",ax=ax, secondary_y=True)
```

# 2.3	Analysis of year-on-year total crime rate

```python
crimes = pd.read_csv('crimes.csv')
crimes.head()
```

```python
population = pd.read_csv('Population.csv')
population.head()
```

```python
req_df = pd.merge(crimes ,population, on=['state'])
req_df["crime_rate"] = (req_df["total_crimes"] / req_df['population'])*100000
req_df
```

```python
fig, ax = plt.subplots(figsize=(25,10))

req_df.plot(x="year_x", y="crime_rate", ylabel = "cases as % of total", ax=ax)
req_df.plot(x="year_y",y="crime_rate",ax=ax, secondary_y=True)
```

Observations: 
    
    1. With population the crime has been increasing over the years
    2. We can see the higher crime rates around 2011-12
